Lets study Indian Restaurants in more details

```python
zomato_India=zomato_data[zomato_data.Country=='India']
zomato_India.shape
```

```python
City_records=zomato_India.groupby('City')['Restaurant ID'].count()
City_records.sort_values(ascending=False).head()
```

```python
City_records.plot(kind='bar',figsize=(10,6))
```

Among the Indian Cities Zomato has major presence in New Delhi followed by Gurgaon,Noida and Faridabad.

```python
cuisine_array=zomato_India.Cuisines.value_counts()
cuisines={}
cnt = 0
for i in cuisine_array.index:
    for j in i.split(', '):
        if j not in cuisines.keys():
            cuisines[j] = cuisine_array[cnt]
        else:
            cuisines[j] += cuisine_array[cnt]
    cnt += 1
```

```python
sorted_cuisines = pd.Series(cuisines).sort_values(ascending=False)
sorted_cuisines.head()
```

```python
plt.figure(figsize=(20,5))
plt.bar(sorted_cuisines[:20].index, sorted_cuisines[:20].values)
```

North Indian is the most popular cuisine. Reason being majority dataset is concentrated towards North Indian region.

```python
bins = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
zomato_India['Aggregate rating_binned'] = pd.cut(zomato_India['Aggregate rating'], bins)
pd.crosstab(zomato_India['Has Online delivery'],columns=zomato_India['Aggregate rating_binned'])
```

```python
pd.crosstab(zomato_India['Is delivering now'],columns=zomato_India['Aggregate rating_binned'])
```
